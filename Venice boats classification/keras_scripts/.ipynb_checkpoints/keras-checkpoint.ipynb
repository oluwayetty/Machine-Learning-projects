```python
import keras
import numpy as np
from keras import backend as K
from keras.models import Sequential
from keras.layers import Activation, Dropout, Flatten, Dense, Input
from keras.optimizers import SGD
from keras.metrics import categorical_crossentropy
from keras.preprocessing.image import ImageDataGenerator
from keras.layers.normalization import BatchNormalization
from keras.layers import MaxPooling2D
from keras.models import Model
# from matplotlib import pylot as plt
from sklearn.metrics import confusion_matrix
import itertools
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
train_path = '../data/venice_traindata' 
validation_path = '../data/venice_valid'
test_path = '../data/venice_testdata'

img_width, img_height = 800, 240
num_train_samples = 4770
num_validation_samples = 2750
epochs = 55
batch_size = 50


if K.image_data_format() == 'channels_first':
    input_shape = (3, img_width, img_height)
else:
    input_shape = (img_width, img_height, 3)
    
print("done")
```

```python
classes = [ 'Alilaguna', 'Ambulanza', 'Barchino', 'Cacciapesca', 'Caorlina', 'Gondola',
           'Lanciafino10m','Lanciafino10mBianca', 'Lanciafino10mMarrone', 'Lanciamaggioredi10m',
           'Lanciamaggioredi10mBianca', 'Lanciamaggioredi10mMarrone', 'Motobarca', 
           'Motopontonerettangolare', 'MotoscafoACTV', 'Mototopo', 'Patanella', 'Polizia',
           'Raccoltarifiuti', 'Sandoloaremi', 'Sanpierota', 'Topa', 'VaporettoACTV', 
           'VigilidelFuoco']
datagen = ImageDataGenerator()

# this is the augmentation configuration we will use for training
gen = ImageDataGenerator(
    rotation_range = 10,
    width_shift_range = 0.1,
    height_shift_range = 0.1,
    shear_range = 0.15,
    zoom_range=0.1,
    channel_shift_range=10,
    horizontal_flip=True)

train_datagen = ImageDataGenerator(
    rotation_range = 10,
    width_shift_range = 0.1,
    height_shift_range = 0.1,
    shear_range = 0.15,
    zoom_range=0.1,
    channel_shift_range=10,
    horizontal_flip=True)

train_batches = train_datagen.flow_from_directory(
    train_path, 
    target_size=(img_width, img_height), 
    class_mode="categorical",
    batch_size= batch_size)

validation_batches = datagen.flow_from_directory(
    validation_path, 
    target_size=(img_width, img_height), 
    class_mode="categorical",
    batch_size= batch_size)

test_batches = datagen.flow_from_directory(
    test_path, 
    target_size=(img_width, img_height), 
    class_mode=None,
    batch_size= batch_size)

print(batch_size)
print(train_batches.class_indices)
```

```python
# plots images with labels within jupyter notebook
# https://github.com/smileservices/keras_utils/blob/master/utils.py
def plots(ims, figsize=(24,12), rows=2, interp=False, titles=None):
    if type(ims[0]) is np.ndarray:
        ims = np.array(ims).astype(np.uint8)
        if (ims.shape[-1] != 3):
            ims = ims.transpose((0,2,3,1))
    f = plt.figure(figsize=figsize)
    cols = len(ims)//rows if len(ims) % 2 == 0 else len(ims)//rows + 1
    for i in range(len(ims)):
        sp = f.add_subplot(rows, cols, i+1)
        sp.axis('Off')
        if titles is not None:
            sp.set_title(titles[i], fontsize=16)
        plt.imshow(ims[i], interpolation=None if interp else 'none')
```

```python
boat_image_path = '../data/venice_traindata/Barchino/20130304_071400_20181.jpg'
image = np.expand_dims(plt.imread(boat_image_path), 0)
plt.imshow(image[0])
```

```python
# Generate batches of augmented image from original image
aug_iter = gen.flow(image)

#Get 12 samples of our augmented image
aug_images = [next(aug_iter)[0].astype(np.uint8) for i in range(12)]
plots(aug_images,figsize=(40,10), rows =3)
```

```python
model = Sequential()
model.add(Conv2D(32, (3, 3), input_shape=input_shape))
# model.add(Conv2D(32, (3, 3), input_shape=input_shape, use_bias= False, bias_initializer='zeros'))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(32, (3, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(64, (3, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Flatten())
model.add(Dense(64))
model.add(Activation('relu'))
model.add(Dropout(0.5))
model.add(Dense(24))
model.add(Activation('softmax'))

model.compile(loss='categorical_crossentropy',
              optimizer='sgd',
              metrics=['accuracy'])

print(input_shape)
model.summary()
```

```python
imgs, labels = next(train_batches)
# plots(imgs, titles = labels)
plots(imgs)
```

```python
print("train steps :" , num_train_samples//batch_size)
print("validation steps :" , num_validation_samples//batch_size)
print("epochs :" , epochs)

model.fit_generator(train_batches, 
                    steps_per_epoch= num_train_samples//batch_size, 
                    validation_data = validation_batches,
                    validation_steps = num_validation_samples//batch_size,
                    epochs = epochs,
                    verbose = 2)
```

```python
predictions = model.predict_generator(test_batches, steps = 197, verbose =2)
print(predictions)
```

```python

# vgg16_model.layers.pop()
model_vgg16_conv = keras.applications.vgg16.VGG16(weights='imagenet', include_top=False)
model_vgg16_conv.summary()

# Make vgg16 model layers as non trainable, i.e freeze the layers to avoid update on weights
for layer in model_vgg16_conv.layers:
    layer.trainable = False

#Create your own input format (here 3x200x200)
input = Input(shape=input_shape,name ='image_input')

#Use the generated model 
output_vgg16_conv = model_vgg16_conv(input)

# Add the flatten layer
x = Flatten(name='flatten')(output_vgg16_conv)
x = Dense(24, activation='softmax')(x)

#Create our own model 
my_model = Model(input=input, output=x)
my_model.summary()

my_model.compile(loss='categorical_crossentropy',
              optimizer='sgd',
              metrics=['accuracy'])
```
